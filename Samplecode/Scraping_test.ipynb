```python
import requests

from bs4 import BeautifulSoup


def scrape_greenhouse_jobs():

    url = "https://boards.greenhouse.io/stripe"

    response = requests.get(url)

    soup = BeautifulSoup(
        response.text,
        "lxml"
    )

    jobs = soup.find_all(
        "table",
        class_="Table_table"
    )

    data = []

    rows = jobs[0].find_all("tr")

    for row in rows:

        cells = row.find_all("td")

        if len(cells) < 3:
            continue

        role = cells[0].text.strip()

        company = "Stripe"

        team = cells[1].text.strip()

        location = cells[2].text.strip()

        link_tag = cells[0].find("a")

        url = (
            link_tag.get("href")
            if link_tag else None
        )

        data.append({
            "role": role,
            "team": team,
            "company": company,
            "location": location,
            "url": url,
            "source": "greenhouse"
        })

    return data
```
